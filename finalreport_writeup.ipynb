## Final Report - ATAC-seq

#### Introduction 

Dendritic cells are key players in the body’s immune response in cancer. One of their main strengths is that they allow for the connection of the innate immune system and the adaptive immune system for tumor immune surveillance purposes [1]. During the development process, dendritic cell progenitors adapt their transcriptional programs to generate subsets of dendritic cells, notably being cDC1 cells, who are responsible for being antigen cross-presenters to CS8 T cells, and cDC2 cells, who bias the immune response to Th2 and Th17 immunity [2]. Advances in research has identified a handful of transcription factors that are important for dendritic cell specification, who are further influenced by epigenetic regulation, most importantly reversible histone acetylation. Alterations in histone acetylation patterns, done by histone acetyltransferases and histone deacetylates (HDACs), have been found to change through different stages of dendritic cell development [3]. Despite this, how chromatic modifications influence development is unclear. The present study aims to identify the role of two epigenetic regulators HDAC1 and HDAC2, within dendritic cell development throughout different stages. To approach this problem, the authors used a variety of bioinformatic techniques. First, the authors utilize cPCR and RNA-sequencing to investigate the role of HDACs in dendritic cell development. The authors also used RNA-sequencing on fluorescence-activated cell sorted cDC1, cDC2, and pDC populations to determine the molecular mechanisms that HDAC1 directly influences. Further, ATAC-seq was employed in order to investigate how HDAC1 is able to regulate chromatin accessibility in the cDC population.

#### Methods

To begin this analysis, the single end reads from the ATAC-seq experiment done mice cells were obtained from the NCBI GEO repository with the accession code GSE266584. The quality of these reads were analyzed first using FastQC v0.12.1 with default parameters [4]. At the same time the GRCm39 mouse reference genome was indexed using the default Bowtie2 v2.5.4 build tool parameters [5]. The adapters and low-quality bases were trimmed from the reads using the single end command with default parameters from Trimmomatic v0.40 [6]. Upon completion, the reads were aligned to the previously indexed mouse reference genome using Bowtie2 v2.5.4 using the ‘--very-sensitive’ flag [7]. Further quality control was then completed using default MultiQC v1.25 [8] and the default flagstat utility from Samtools v1.22.1 [9]. If any of the aligned reads matched aligned with mitochondrial DNA, they were removed using default samtools v2.23.1 parameters [10]. The remaining alignment files were then sorted in genomic order based on their alignment to the reference genome and indexed using Samtools v1.22.1 with default parameters for further analysis [11].


Bigwig files were generated from the previously sorted BAM files using default deeptools v3.5.5 and signal coverage plots were calculated using deeptools’s computeMatrix utility with ‘–referencePoint center’ specified for ATAC-seq purposes [12].  These plots included 1500 bp cushions on either side of the transcription start site and end site. A spearman correlation analysis was also performed with deeptools v3.5.5 utilities plotcorrelation and multibwsummary with default parameters to determine the similarity between samples [13]. Using the output from deeptools’ computeMatrix, the TSS Enrichment score was calculated using a custom script that was sent through deeptools [14]. The fraction of reads in peaks score  were calculated using a small custom pipeline in which total mapped reads were counted with default parameter samtools [15].


Peak calling was executed using the MACS3 suite v3.0.3 using the ‘—nomodel –keep-dup auto -q 0.01’ parameters [16]. Any peaks that overlapped in undesirable regions as denoted by the mm38-blacklist v2 as poised by the Boyle lab (https://github.com/Boyle-Lab/Blacklist/tree/master/lists) were removed through the default use of bedtools v2.31.1 [17]. Peak annotation and motif finding was done on the set of filtered peaks using the HOMER suite v4.11 with default options [18]. Motif finding was done on the set of peaks using the HOMER suite v4.11 with the specified parameters ‘-size 200 -mask’ [19]. These peaks were lastly analyzed using Diffbind v3.20.0 with default parameters to identify statistically significant differences in accessibility between the two experimental conditions [20].


#### Sequencing Quality Control 

Looking through the multiqc report shows that the sequences that were obtained by the ATAC-seq experiment are suitable for downstream analysis. First, the GC content of each sample is reported to be normally distributed around 50%. This a good sign that indicates that the samples are of high quality. Next, the report displays that the Phred scores of each sample starting off around 30 and rise to be around 40, which is maintained throughout the length of the sequences. This also indicates that the samples are of high quality and that mean quality scores for every sample is considered as very good. Further, the report concludes that the per base N content of all 8 samples is 0%, with it peaking just before 1% near the beginning of the sequences. This means that the sequencer was able to confidently identify the present nucleotides, further indicating that the reads were of high quality. The quality check flags two overrepresented sequences but as they only account for 0.06% of all reads at most, they can safely be ignored. The general statistics section displays that there is a slight difference in the samples sequencing depth. All raw samples that correspond with the cDC1 cells have a consistent sequencing depth of around 30M, while the raw samples that correspond with the cDC2 cells have a sequencing depth that ranges from 22.4M to 31M. However, this section also reports that each sample has about 0.5% of reads being dropped, which is another good sign that these sequences are of good quality. Additionally, it also shows that each sample has a very low amount of sequence duplications, with the mean being around 18.15%. The report also shows that raw reads are contaminated with adapter sequences. However, this contamination is removed in the pipeline using trimmomatic.

#### Alignment Statistics

The alignment statics were also calculated with Samtools flagstat and displayed within the multiqc report. Across all samples, the flagstat also reports that the mapping rates are high quality, with about 19.5M to 29M reads out of about 24.5M to 33M reads being successfully aligned to the reference genome. This passes the ATAC-seq threshold of about 70% of successfully aligned reads. Flagstat also reports that duplications levels are extremely low, supporting the idea that the ATAC-seq library has no PCR amplification bias in the representation of open chromatin regions and will more realistically represent true biological signal. These metrics are generally seen to be consistent between the samples, which ensures that any observed differences will reflect biological variation and not technical variation. Overall, this report supports that the aligned data is able to be used for differential accessibility analysis between cDC1 and cDC2 cell types.

#### ATAC-seq Quality Control Metrics - TSS Enrichment Scre & FRiP

The first ATAC-seq specific quality control metric calculated for these reads was the TSS enrichment score of each sample. This metric measures the signal to noise ratio of each sample through comparison of DNA accessibility or binding at actual gene start sites vs the background regions. The TSS enrichment score that was calculated for the cDC1 samples to have an average of 14.388, where each were score was found to be 13.570, 13.826, 15.485, and 14.388.  The TSS enrichment score that was calculated for the cDC2 samples were slightly smaller with an average of 13.432, where each were score was found to be 13.54, 14.492, 12.156, and 13.393.  These scores are considered as excellent. They indicate that the ATAC-seq library was enriched for biologically relevant open chromatin, supporting the fact the idea that the experiment was successful. The second ATAC-seq specific quality control metric calculated was the fraction of reads in peaks (FRiP). Alongside measuring the signal-to-noise ratio and the enrichment quality of the data, this metric measures the percentage of total sequenced reads that fall within identified peak regions.  For this experiment, the FRiP for the cDC1 cells had an average of 0.23, where the KO conditions had scores of 0.2138 and 0.2715 and the WT condition had scores of 0.2256 and 0.2046. The cDC2 cells had an overall average score of about 0.24, where the KO conditions had score of 0.2538 and 0.2819 and the WT conditions had scores of 0.2745 and 0.2104. Each of these samples meet the minimum threshold FRiP score of 0.20, supporting the idea that the experiment can be considered as successful and that any peaks that were found can be considered real and enriched.

#### Differentially Accessible Regions the Pipeline Discovered

This pipeline was able to identify differentially accessible regions in both cell conditions. Regarding the cDC1 cells, this pipeline was able to identify a total of 1619 peaks using a significance threshold of p < 0.01. More specifically, it was able to determine that 1250 peaks in the knockout condition had gained accessibility and had a positive fold change. It also found that 369 peaks in the knockout condition had lost accessibility and reported a negative fold change. Regarding the cDC2 cells, this pipeline was able to identify 751 differentially accessible regions using a significance threshold of p < 0.01. It further determined that 539 of these peaks in the knockout condition had gained accessibility and reported a positive fold change. It also showed that 212 of these peaks in the knockout condition had lost accessibility and reported a negative fold change.

Completing gene ontology analysis on the enriched regions of the differentially accessible regions gives the results displayed in figure 1 and 2 above. This analysis shows that there are many similar processes that are being enriched in both cDC1 and cDC2 cells. These processes are related to cytoskeletal development, dendrite development, and neuronal projection, which are consistent with the identity of the cells and their known functions. However, this analysis reveals some differences in the enrichments between the cell types. For example, cDC1 cells show enrichment terms that are associated with small GTPase-mediated signaling, actin filament organization, and cell junction assembly as seen in figure 1. These suggest that changes in the chromatin accessibility of the cDC1 cells could also have implications in cellular structure and signaling processes. In contrast, the enrichment results of the cDC2 cells show that there are additional enrichment in processes such as mesenchyme development, developmental growth, cell migration–related processes, immune cell differentiation and hematopoiesis as seen in figure 2. These results suggest that the difference in chromatin accessibility in cDC2 cells could also be linked to differentiation and migratory behavior.


#### Motif Enrichment

<img src="pics/motif_cdc1.png" alt="Motif Enrichment for cDC1 Cells" width="800"> 

Fig 3: Top 5 motifs discovered in cDC1 cells for both WT and KO conditions

Shown in figure 3 above are the top 5 motifs in the cDC1 samples that were identified by HOMER and compiled in the knownResults output. The top portion of the figures displays the motifs found in the WT condition and the bottom portion of the figures displays those found in the KO condition. Some of the top enriched motifs found in the WT condition include members of the ETS family of transcriptions factors who play a role in dendritic cell development, including PU.1, SPIB, ELF5, and ELF4. It also has IRF8, which is an interferon regulatory factor that is also associated with dendritic cell development. The motifs that are enriched in the KO conditions are the same as to those found in the WT condition. This suggests that the knockout influences the activity or targeting of core dendritic cell transcription factors at specific loci rather than fundamentally altering the cell's regulatory architecture. Some additional motifs that arise in the top 10 enriched list include PU.1:IRF, which reinforces the importance of cooperative binding of transcription factors as differentially accessible sites, and PU.1:IRF8, which is the heterodimer binding site where the two proteins bind together again reinforcing the importance of cooperative binding of master regulators that are key for dendritic cell development.

Fig 4: Top 5 motifs discovered in cDC2 cells for both WT and KO conditions

Shown in figure 4 above are the 5 motifs in the cDC2 samples that were identified by HOMER and compiled in the knownResults output. The top portion of the figures displays the motifs found in the WT condition and the bottom portion of the figures displays those found in the KO condition. Some of the top enriched motifs found in both conditions include PU.1, SPIB, ELF4, and ELF5, which are all members of the ETS transcription factor family, which is known for its essential role in dendritic cell development. There is also CTCF, which is a chromatin architectural protein. The only difference between the two is that CTCF is much more enriched in the knockout condition, suggesting that the knockout may be affecting chromatin organization and 3D genome structure alongside transcription factor binding. Some other interesting motifs that show up in the top 10 list is BORIS, a paralog of CTCF that ranks higher in the KO condition, as well as EHF, which is an epithelial-specific ETF factor that is unexpected in dendritic cells, suggesting that there could be cross-lineage transcriptional signatures within these immune cells.

The opposite is true of their loss regions in their respective WT conditions. ##### Scatter plots 

<img src="pics/scatter.png" alt="Recreation of 6c and 6e" width="800"> 

Fig 6: Correlations between gene expression and differential accessbile regions in cDC1 and cDC2 cells

Figure 6 shows a scatter plot that compares the changes in gene expression and differential accessible regions for cDC1 and cDC2 cells, specifically highlighting the gain of accessibility regions and increased gene expression from data analyzed in this analysis. For cDC1 cells, this analysis shows that the genes that have an increase in both gene expression and chromatin accessibility include Spib, Susd2, Nectin2, Csf3r, Irf8, Celf4, and Maged1. This is deviates from what is shown in the paper, which reports that only Cd8a, Maged1, and Ccr2 are the genes in cDC1 cells that have this increase in both gene expression and chromatin accessibility.  Despite this, both analyses report that Maged1 has this increase. Additionally, the overall distribution of both the up and down regulated genes is comparable between the paper and this analysis. For cDC2 cells, this analysis revealed that Spib, Susd2, Nectin2, Csf3r, Irf8, Celf4, and Maged1 also display this increase. This is largely consistent with the findings of the paper, which reports that Nectin2, Csf3r, Celf4, and Spib also exhibit this increase in gene expression and chromatin accessibility. Additionally, the distribution of the rest of the upregulated and downregulated genes from this analysis visually match to what was reported in the paper. Taken together, these observations suggest that the cDC2 results are more closely replicated than those observed for cDC1.

#### Future Directions

One additional question that could be interesting to address would be: “Can modulation on HDAC1 activity combine with existing cancer immunotherapies to improve treatment efficiency?” This would be relevant as combining epigenetic modulators with immunotherapies could make cancer treatment more efficient and less harsh on the body in comparison to current treatment methods. A second question could be: “How does HDAC1 influence the epigenetic regulation of key transcription factors during dendritic cell differentiation?” This would be relevant because the motif enrichment analysis with cDC1 cells highlighted that there may is some importance between cooperative binding between transcription factors and subsequent investigation would reveal how immune cell identity and function could be controlled at the chromatin level. A third question that can be posed is: “What are the long-term effects of HDAC1 modulation on systemic immune homeostasis and autoimmunity?” This is relevant as altering key epigenetic regulators like HDAC1 could have an impact beyond anti-tumor immunity.

#### Conclusions

To conclude, this analysis shows that the loss of HDAC1 leads to cell-type specific changes in chromatin accessibility in cDC1 and cDC2 cells. This pipeline was able to identify a substantial amount of differentially accessible regions, with a predominance of regions gaining accessibility in the KO condition. These were generally consistent with HDAC1’s role as a transcriptional repressor. Functional enrichment analysis revealed that there are shared biological themes across both cell types, such as cytoskeletal organization and dendrite related processes, as well as cell-specific biological themes, such as cDC1s associations with signaling and structural organization and cDC2s associations with cell migration and hematopoietic processes. Motif enrichment analyses further showed that core dendritic cell transcription factors, particularly ETS family members and IRF-associated motifs, remain central in both wild-type and knockout conditions, suggesting that HDAC1 loss alters chromatin accessibility at established regulatory sites as opposed to altering the molecular population of transcription factors. Lastly, reproducing the figures from the original study showed that the results found from this pipeline agree with the findings of the paper. When combined, all of these findings support the conclusion that HDAC1 plays a critical role in modulating chromatin accessibility to regulate dendritic cell-specific transcriptional programs, allowing for subtype-specific functional outcomes.
